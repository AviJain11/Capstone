```python
#importing common libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Extracting information from the wikipedia URL
#Tried multiple methods, found this to work best.. it also use bs4 refernce of beautiful soup

```python
data = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', flavor='bs4', header=0)
df = data[0]
```

```python
df.head()
```

```python
#Ignoring Borough = Not assigned
df.drop(df[df['Borough'] == 'Not assigned'].index, inplace=True)
```

```python
df.head(15)
```

```python
df['Neighbourhood'] = df.groupby(['Postcode','Borough'])['Neighbourhood']\
       .apply(lambda x: x.cumsum())
df.head()
```

```python
#Dropping duplicate values
df.drop_duplicates(subset=['Postcode','Borough'], keep='last', inplace=True)
```
